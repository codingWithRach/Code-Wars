In a room is a table with a pile of cakes. One of these cakes is poisoned. Every turn, we will take cakes from this pile and eat them, leaving the last one to be the poisoned cake. Whoever eats the poisoned cake will die. The poisoned cake is clearly marked, so you do not have to guess. Rather you need to rely on logic to save you.

# The rules are simple:

1. Do not eat the last cake. It's obivious that the poisoned cake will be the last to be eaten, so, DON'T EAT THE LAST CAKE. Try to make your opponent eat it.

2. When it is your turn, you can only take one, two or three cakes. The same rules apply to your opponent on their turn. You cannot skip your move, so choose wisely how many cakes to eat. Both opponents will be able to see how many cakes the other eats on each turn.

3. You cannot copy your opponent's previous move; likewise they cannot copy yours. If your opponent takes one cake, next move you can only choose between two and three. If you take three cakes, your opponent can only choose one or two. This doesn't effect the first move, only subsequent ones.

4. If one of the players has no valid moves (ie. one cake left and previous move was one cake), that player will skip their turn. Then the other player will be forced to eat the last cake. This is the ONLY case of turn skipping.

5. You can choose whether or not to go first. This decision is key to victory, so don't hurry, choose wisely!

# My notes
If cakes % 4 == 2 we cannot win, so need to avoid this state (and force our opponent into it) e.g.
- if there are 6 left and you choose 1 (leaving 5), opponent will take 3 (leaving 2) - you can either take 1 or 2 but will lose either way
- if there are 6 left and you choose 2 (leaving 4), opponent will take 3 (leaving 1, which you have to take)
- if there are 6 left and you choose 3 (leaving 3), opponent will take 1 (leaving 2, which you have to take)

So if the starting number of cakes % 4 == 2 (or there's only 1 left), ensure your opponent goes first.  Otherwise you go first.

If cakes % 4 == 0 you can choose 3 cakes, leaving cakes % 4 == 1 - opponent will either take 1 (resulting in cakes % 4 == 0 again) or 2 (resulting in cakes % 4 == 3 - see below)

If cakes % 4 == 1 you can choose 3 cakes, leaving cakes % 4 == 2 i.e. opponent cannot win

If cakes % 4 == 3 you can choose 1 cake, leaving cakes % 4 == 2 i.e. opponent cannot win

```python
class Player:
  def __init__(self, cakes):
    pass

  # Decide who moves first
  # If cakes % 4 == 2 we cannot win so we need to make sure the opponent goes first
  # Otherwise, we go first
  def firstmove(self, cakes):
    if cakes == 1 or cakes % 4 == 2:
      return False
    return True

  # Decide your next move
  def move(self, cakes, last):
    # if number of cakes remaining is a multiple of 4 then choose 3 cakes, leaving cakes % 4 == 1 - opponent will either take 1 (resulting in cakes % 4 == 0 again) or 2 (resulting in cakes % 4 == 3 - see below)
    if cakes % 4 == 0:
      return 3

    # if number of cakes remaining is 1 more than a multiple of 4 then choose 3 cakes, leaving cakes % 4 == 2 (a position from which the opponent cannot win)
    if cakes % 4 == 1:
      return 3

    # if number of cakes remaining is 1 less than a multiple of 4 then choose 1 cake, leaving cakes % 4 == 2 (a position from which the opponent cannot win)
    if cakes % 4 == 3:
      return 1

    # should never have cakes remaining being a multiple of 4 + 2
    return 0
```
