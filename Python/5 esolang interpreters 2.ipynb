```python
def interpreter(code, tape):
    tape_list = list(tape)
    pointer_pos = 0
    code_pos = 0
    while code_pos < len(code):
        char = code[code_pos]

        # move pointer to left by 1 cell - terminate program if pointer out of bounds
        if char == '<':
            pointer_pos -= 1
            if pointer_pos < 0:
                return ''.join(tape_list)

        # move pointer to right by 1 cell - terminate program if pointer out of bounds
        elif char == '>':
            pointer_pos += 1
            if pointer_pos >= len(tape):
                return ''.join(tape_list)

        # flip bit at current cell
        elif char == '*':
            if tape_list[pointer_pos] == '0':
                tape_list[pointer_pos] = '1'
            else:
                tape_list[pointer_pos] = '0'

        # jump past matching ] if value at current cell is 0
        elif char == '[':
            if tape_list[pointer_pos] == '0':
                pos = code_pos + 1      # position immediately after current [
                bracket_count = 1       # we're looking for the next ] until proved otherwise
                found = False
                # move forward through remaining characters in the code
                while pos < len(code) and not found:
                    # if we find a ] then reduce the count of ] we're looking for.  Once we reach 0 then this is the matching ]
                    if code[pos] == ']':
                        bracket_count -= 1
                        if bracket_count == 0:
                            # matching bracket
                            code_pos = pos
                            found = True
                    # if we find another [ then we need to find another ] before we reach the matching ]
                    elif code[pos] == '[':
                        bracket_count += 1
                    # move to the next character
                    pos += 1
                # if there's no matching ], the pointer is out of range
                if not found:
                    return ''.join(tape_list)

        # jump back to matching [ if value at current cell is non-zero
        elif char == ']':
            if tape_list[pointer_pos] != '0':
                pos = code_pos - 1      # position immediately before current ]
                bracket_count = 1       # we're looking for the next [ until proved otherwise
                found = False
                # move back through all previous characters in the code
                while pos >= 0 and not found:
                    # if we find a [ then reduce the count of [ we're looking for.  once we reach 0 then this is the matching [
                    if code[pos] == '[':
                        bracket_count -= 1
                        if bracket_count == 0:
                            # matching bracket 
                            code_pos = pos 
                            found = True
                    # if we find another ] then we need to find another [ before we reach the matching [
                    elif code[pos] == ']':
                        bracket_count += 1
                    # move to the next character
                    pos -= 1
                # if there's no matching [, the pointer is out of range]
                if not found:
                    return ''.join(tape_list)
                
        # move to the next character in the code
        code_pos += 1
    return ''.join(tape_list)
```
